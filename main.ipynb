```python
import folium
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
muni = gpd.read_file(filename='muni.geojson')
muni.head()
```

```python
instruction_zones = gpd.read_file(filename='instructions.geojson')
instruction_zones.head(30)
```

```python
intersection_muni = gpd.sjoin(left_df=muni[['Muni_Code','geometry']],right_df=instruction_zones[['OBJECTID','geometry']],how='inner')
intersection_muni.head()
```

```python
intersection_muni_gb= (intersection_muni.groupby('Muni_Code').count()['OBJECTID'])
ls_muni_intersect =list(intersection_muni_gb.loc[lambda t:t>1].index)
ls_muni_intersect
```

```python
intersection_muni = intersection_muni[intersection_muni['Muni_Code'].isin(ls_muni_intersect)]
```

```python
# find the the first intersection
instruction_zones[ instruction_zones['merchav'].str.contains('נגב') ]
```

```python
negev=intersection_muni[intersection_muni['OBJECTID'].isin([31,16])]
negev
```

```python
from folium import DivIcon


cnter_negev= muni[muni['Muni_Code'].isin(negev['Muni_Code'])]['geometry'].centroid.get_coordinates()
# print(cnter_negev)
m =folium.Map(location=[cnter_negev.mean()['y'],cnter_negev.mean()['x']],zoom_control=False,zoom_start=10)
folium.GeoJson(instruction_zones[instruction_zones['OBJECTID'].isin([16,31])],  style_function=lambda feature: {
        "fillColor": "#20A12B"
        if feature['properties']['Instruction_Policy'] ==1
        else '#F0F235'if feature['properties']['Instruction_Policy'] ==2
        else '#E1A817' if feature['properties']['Instruction_Policy'] ==3
        else 'green' ,
        "color": "black",
        "weight": 2,
        "opacity":0.9,
        "fillOpacity":0.8
    }).add_to(m)

# add labels
for inst in [16,31]:
    folium.map.Marker(
    [instruction_zones[instruction_zones['OBJECTID']==inst]['geometry'].centroid.get_coordinates()['y'], instruction_zones[instruction_zones['OBJECTID']==inst]['geometry'].centroid.get_coordinates()['x']],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-family:david;font-size: 15pt">%s</div>' % str(instruction_zones[instruction_zones['OBJECTID']==inst]['merchav'].values[0]),
        )
    ).add_to(m)
folium.GeoJson(muni[muni['Muni_Name'].isin(['מרחבים','אשכול','בני שמעון'])],  style_function=lambda feature: {
        "fillColor": "gray",
        "color": "black",
        "weight": 0.5,
        "fillColor": "gray",
        "fillOpacity":0.6
    }).add_to(m)
for inst in ['מרחבים','אשכול','בני שמעון']:
    folium.map.Marker(
    [muni[muni['Muni_Name']==inst]['geometry'].centroid.get_coordinates()['y'], muni[muni['Muni_Name']==inst]['geometry'].centroid.get_coordinates()['x']],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-weight:bold;font-family:david;font-size: 9pt">%s</div>' % str(muni[muni['Muni_Name']==inst]['Muni_Name'].values[0]),
        )
    ).add_to(m)
m
```

```python
from folium.plugins import FloatImage
url = ( 
    "https://raw.githubusercontent.com/ocefpaf/secoora_assets_map/"
    "a250729bbcf2ddd12f46912d36c33f7539131bec/secoora_icons/rose.png"
)
FloatImage(url,bottom=57, left=75).add_to(m)
m
```

```python
import io 
from PIL import Image
img_data = m._to_png(5)
img = Image.open(io.BytesIO(img_data))
img.save('image.png')
```
