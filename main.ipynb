```python
import folium
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
muni = gpd.read_file(filename='muni.geojson')
muni.head()
```

```python
instruction_zones = gpd.read_file(filename='instructions.geojson')
instruction_zones.head()
```

```python
intersection_muni = gpd.sjoin(left_df=muni[['Muni_Code','geometry']],right_df=instruction_zones[['OBJECTID','geometry']],how='inner')
intersection_muni.head()
```

```python
intersection_muni_gb= (intersection_muni.groupby('Muni_Code').count()['OBJECTID'])
ls_muni_intersect =list(intersection_muni_gb.loc[lambda t:t>1].index)
ls_muni_intersect
```

```python
intersection_muni = intersection_muni[intersection_muni['Muni_Code'].isin(ls_muni_intersect)]
```
